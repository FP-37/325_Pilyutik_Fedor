Сортировка методом вставок работает путем последовательного добавления элементов в уже отсортированную часть массива. Внешний цикл проходит по элементам массива, начиная с индекса 1 до n-1 (так как первый элемент уже считается отсортированным). Для каждого элемента на позиции i текущему элементу key присваивается значение array[i].
Переменная j инициализируется значением i-1, что указывает на предыдущую позицию в массиве. Внутренний цикл while проверяет, является ли текущий элемент больше key. Если это условие выполняется, элемент сдвигается вправо, пока не будет достигнута позиция, где array[j] меньше или равно key. Элементы массива от j+1 до 0 (если j больше или равно 0) сдвигаются на одну позицию вправо. j уменьшается на 1 на каждой итерации сдвига. После завершения внутреннего цикла, текущий элемент key вставляется на правильную позицию, заменяя сдвинутый элемент array[j+1].
Преимущество такого способа сортировки в низком количестве перестановок.

```python
def main():

    n = int(input().strip())
    array = list(map(int, input().strip().split()))

    for i in range(1, n):
        key = array[i]
        j = i-1
        while j >= 0 and array[j] > key:
            array[j+1] = array[j]
            j -= 1
        array[j+1] = key

    print(*array)
```
