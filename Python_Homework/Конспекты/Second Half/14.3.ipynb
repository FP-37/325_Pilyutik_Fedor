Функция min_fives(a, b, c) определяет минимальное количество оценок "5" (left), которые нужно добавить к существующим оценкам a (оценок "2"), b (оценок "3") и c (оценок "4"), чтобы средний балл группы стал не меньше 4 (с округлением до ближайшего целого). Используется бинарный поиск в диапазоне от 0 до суммы всех оценок. Функция is_valid проверяет, удовлетворяет ли текущее количество оценок "5" (mid) условию: средний балл, вычисленный с учетом добавленных оценок "5", после округления до ближайшего целого, больше или равен 4. Если условие выполняется, мы уменьшаем правую границу поиска, иначе увеличиваем левую. В итоге функция возвращает минимальное количество оценок "5", необходимых для достижения целевого среднего балла.

```python
def min_fives(a, b, c):
    total_grades = a+b+c
    target_average = 4

    def is_valid(num_fives):
        total_score = a*2+b*3+c*4+num_fives*5
        average_score = total_score/(a+b+c+num_fives)
        rounded_average = int(average_score+0.5)
        return rounded_average >= target_average

    left, right = 0, total_grades
    while left <= right:
        mid = (left+right)//2
        if is_valid(mid):
            right = mid-1
        else:
            left = mid+1

    return left
```

Введём количество оценок поочерёдно, используем нашу функцию и результат её выполнения выведем на экран.

```python
def main():
    a = int(input().strip())
    b = int(input().strip())
    c = int(input().strip())

    print(min_fives(a, b, c))

if __name__ == "__main__":
    main()
```

Выполненная задача:

В лицее на уроках информатики ответы учеников оцениваются целым числом баллов от 2 до 5.
Итоговая оценка по информатике выставляется как среднее арифметическое оценок на всех уроках,
округленное до ближайшего целого числа. Если среднее значение находится ровно посередине между двумя целыми числами,
то оценка округляется вверх.

Все ученики лицея стремятся получить итоговую оценку по информатике не ниже 4 баллов.
К сожалению, один из учеников получил на уроках a двоек, b троек и c четверок.
Теперь он планирует получить несколько пятерок, причем хочет, чтобы итоговая оценка была не меньше 4 баллов.
Ему надо понять, какое минимальное количество пятерок ему необходимо получить, чтобы добиться своей цели.

Требуется написать программу, которая по заданным целым неотрицательные числам a, b и c определяет
минимальное количество пятерок, которое необходимо получить ученику, чтобы его итоговая оценка по информатике
была не меньше 4 баллов.

Входные данные
Входные данные содержат три строки. Первая строка содержит целое неотрицательное число a,
вторая строка содержит целое неотрицательное число b, третья строка содержит
целое неотрицательное число c (0≤a,b,c≤10^15,a+b+c≥1).

Выходные данные

Выходные данные должны содержать одно число: минимальное число пятерок, которое необходимо получить ученику,
чтобы итоговая оценка была не меньше 4 баллов.
