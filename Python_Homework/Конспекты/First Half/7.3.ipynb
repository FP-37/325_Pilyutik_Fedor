Из примечательных нововведений среди нашего кода в данной задаче реализовано использование функции items() для словарей. Путём использования такой функции мы можем вернуть не только каждое из значений словаря в цикле, но и указанные ключи, что является принципиально важным при выполнении условий выполняемой задачи.
Укажем также некоторые другие функции для работы со словарями:
keys() — Возвращает список всех ключей в словаре.
values() — Возвращает список всех значений в словаре.
items() — Возвращает список кортежей, содержащих пары “ключ-значение” (дополнение информации, что функция возвращает список кортежей).
get(key, default) — Возвращает значение по указанному ключу. Если ключ отсутствует, возвращает значение default.
setdefault(key, default) — Устанавливает значение по ключу, если ключ отсутствует. Возвращает текущее значение по ключу или значение default, если ключ был добавлен.
update(other) — Обновляет словарь текущими парами “ключ-значение” из другого словаря.
pop(key) — Удаляет элемент с указанным ключом и возвращает его значение. Если ключ отсутствует, возвращает None.
popitem() — Удаляет и возвращает случайный элемент словаря (пару “ключ-значение”).
clear() — Удаляет все элементы из словаря.

В самой программе сначала реализуем ввод через клавиатуру указанного количества пар синонимов, поместим все пары в словарь - пусть первое слово будет ключом, а второе значением. А далее будем проверять, входит ли введённое последним слово в набор ключей или значений, и соотвественно выводим либо нужный ключ, либо нужное значение при нахождении совпадения.

```python
def main():

    n = int(input())
    word_pairs = {}
    for _ in range(n):
        pair = input().split()
        word_pairs[pair[0]] = pair[1]
    user_input = input()
    for k, v in word_pairs.items():
        if v == user_input:
            print(k)
        elif k == user_input:
            print(v)

if __name__ == "__main__":
    main()
```

Выполненная задача:

Вам дан словарь, состоящий из пар слов. Каждое слово является синонимом к парному ему слову.
Все слова в словаре различны. Для одного данного слова определите его синоним.

Входные данные
Программа получает на вход количество пар синонимов N. Далее следует N строк,
каждая строка содержит ровно два слова-синонима. После этого следует одно слово.

Выходные данные
Программа должна вывести синоним к данному слову.

Примечание
Эту задачу можно решить и без словарей (сохранив все входные данные в списке),
но решение со словарем будет более простым.
