Нам необходимо, чтобы при выводе числа слева-направо и сверху-вниз в нашей матрице шли последовательно. Во избежание обнулений значений и правильного заполнения списка сделаем следующим образом - пусть логика заполнения значений элементов массива работает так, что мы каждый раз будем считывать индекс столбца и к нему прибавлять произведение индекса строки и её длины.

```python
def main():
    inpt = input().strip().split()
    n = int(inpt[0])
    if len(inpt) < 2:
        for row in [[j + i * n for j in range(n)] for i in range(n)]:
            print(*row)
    else:
        m = int(inpt[1])
        for row in [[j + i * m for j in range(m)] for i in range(n)]:
            print(*row)

if __name__ == "__main__":
    main()
```
