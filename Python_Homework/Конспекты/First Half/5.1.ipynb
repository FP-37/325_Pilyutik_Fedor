Задание требует только создать генератор, но также я написал код для воссозданий полных условий задачи ради перепроверки себя. В переменную inpt помещаем либо число, либо пару чисел, разделённых пробелом. В зависимости от количества введённых чисел сработает либо генератор для квадратной матрицы, либо для матрицы случайного размера.

Принцип работы генераторов в Python:
Генератор может быть реализован либо как отдельная функция, в которой визуальное отличие будет заключаться в использовании ключевого слова yield вместо return, либо как генераторное выражение, что мы и реализуем в коде ниже.
Функциональная особенность генераторов заключается в возвращении объекта-генератора - несохраняемых в памяти данных, которые мы можем использовать в моменте, к примеру, для вывода значений в консоль. Они будут полезны для уменьшения используемой памяти при работе с большими данными.

Реализация генераторных выражений в коде ниже следующая:
Рассмотрим девятую строку. Указанный генератор создает двумерный список, в котором каждый объект представляет собой произведение индексов строк и столбцов - i*j - проход по каждому индексу реализуем через два цикла for с указанием идти от 0 до n и от 0 до m, где m и n - это заданные изначально размеры двумерного списка. Далее вытаскиваем каждый одномерный список из нашего генератора и печатаем их по очереди.

```python
def main():
    inpt = input().strip().split()
    n = int(inpt[0])
    if len(inpt) < 2:
        for row in [[i * j for j in range(n)] for i in range(n)]:
            print(*row)
    else:
        m = int(inpt[1])
        for row in [[i * j for j in range(m)] for i in range(n)]:
            print(*row)
```

Не забываем вооот об этой части (см. 3.1). В дальнейшем сниму акцент с этой части кода, чтобы отдать приоритет в конспектах другим аспектам.

```python
if __name__ == "__main__":
    main()
```
