Реализуем функцию, которая будет принимать в качестве входящих переменных введённый список данных и число x, количество совпадений с которым мы считаем в списке. Внутри самой функции реализуем счётчик совпадений - изначально присвоим ему значение 0. Также реализуем цикл, в котором будем сравнивать каждый элемент списка с заданным значением x и при наличии совпадения увеличивать счётчик. Из функции выводим целочисленное значение счётчика.

```python
def count_occurrences(array, x):
    count = 0
    for num in array:
        if num == x:
            count += 1
    return count
```

Реализуем функцию main, в которой будет содержаться основная часть нашей программы.
Во второй строке реализуем ввод числа X, количество совпадений с которым будем искать. Для предупреждения случайных пробелов перед числом пропишем функцию strip() для нашего ввода, а также используем функцию int(), чтобы сразу сменить тип введённых данных со строчного на целочисленный.

В третьей строке реализуем ввод элементов массива - для введённой строки сразу используем функцию strip() для удаления случайных пробелов перед вводом элементов, далее используем функцию split() для преобразования нашей строки в список - по умолчанию, без указания сепаратора, split() разобьёт строку на список по пробелам внутри строки.
Далее в той же строке обернём полученный список в функцию map(<функция>,<массив данных>) - данная функция передаёт каждый элемент указанного во втором параметре массива данных указанной в первом параметре функции и обрабатывает в соответствии с опциональностью этой функции.
map() вернёт объект карты, поэтому используем функцию list() для преобразования полученного объекта обратно в список. Положим его в переменную array.

В четвёртой строке аналогично второй - введём новое число. Однако инициализированную переменную x будем использовать в качестве числа, повторы к которому будем подсчитывать. Переменную N же использовали для указания количества элементов массива, но при указанном в примерах тестов задачи вводе её наличие рудиментарно - можно убрать. Или же найти ей применение в качестве ограничителя по количеству элементов и дописать проверку на количество и последующее удаление лишних.

В пятой строке инициализируем переменную result, которой присвоим выходящее значении функции count_occurrences, принявшей наш список и искомое для подсчёта число. В шестой выведем результат на экран - можно пожертвовать читаемостью и сразу поместить в print() функцию, избежав переменной result, но пусть код останется более читаемым.

```python
def main():
    N = int(input().strip())
    array = list(map(int, input().strip().split()))
    x = int(input().strip())
    result = count_occurrences(array, x)
    print(result)
```

В данной задаче применим ниже указанную конструкцию. Она является элементом реализации ООП в Python - в данном случае наша программа компилируется только в случае, если скрипт запускается напрямую, и избегает случайного запуска при импорте.
__name__ - это атрибут, которому присваивается имя модуля. Имя будет разным в зависимости от контекста:
Скрипт запущен напрямую: Если скрипт выполняется как основная программа, __name__ устанавливается в значение "__main__".
Модуль импортирован: Если модуль импортируется другим модулем, __name__ принимает имя модуля.
Скрипт запущен в интерактивном режиме: Если скрипт запущен в интерактивном режиме, __name__ также будет именем модуля.

```python
if __name__ == "__main__":
    main()
```
