Опять же, открываем файл input.txt в режиме чтения, создаём словарь для подсчёта, а далее смотрим на отдельные строки файла, так как знаем, что в каждой строке указаны только два значения и при этом в упорядоченном виде - сначала фамилии кандидатов, потом количество голосов.
В словаре под ключами будем подразумевать фамилии кандидатов, а под значениями - количество голосов как целочисленное значение. При возникновении каждой новой строки увеличиваем значение в словаре под указанным ключём на необходимое количество голосов. Далее сортируем ключи с помощью sorted() - по умолчанию для списка строк как раз фамилии кандидатов отсортируются в лексикографическом порядке. В выводе реализуем пары, состоящие из фамилий кандидатов из отдельного списка и сопутствующие значения словаря.

```python
def main():

    with open('input.txt', 'r') as file:
        candidate_votes = {}

        for line in file:
            candidate, votes = line.strip().split()
            candidate_votes[candidate] = candidate_votes.get(candidate, 0) + int(votes)

        candidates = sorted(candidate_votes.keys())

        for candidate in candidates:
            print(candidate, candidate_votes[candidate])

if __name__ == "__main__":
    main()
```
