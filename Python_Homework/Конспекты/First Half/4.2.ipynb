Другой способ создания двумерных списков - создание списка через генератор с заполнением его списками путём реализации внутреннего цикла. Более подробно о генераторах см. 5.1. В данном случае инициализированный одномерный список array заполняется одномерными списками n раз, при том в каждом вложенном списке будет n элементов. Таким образом получится квадратная матрица.
Далее определим среднюю строку и полностью заменим все элементы строки на *.
Также поступим со средним столбцом, после чего останется заполнить диагонали.
В главной диагонали номера заполняемых столбцов совпадают с номерами строк, в побочной номера столбцов будут наоборот - от конца к началу.
Далее выведем рисунок на экран - чтобы отобразить наш список в матричном виде используем цикл.
Функция ' '.join() объединяет списки строк в одну строку с указанным разделителем - в данном случае с пробелом.

```python
def main():
    n = int(input().strip())
    array = [['.'] * n for _ in range(n)]
    mid_row = n // 2
    mid_col = n // 2
    for i in range(n):
        array[mid_row][i] = "*"
        array[i][mid_col] = "*"
        array[i][i] = "*"
        array[i][n - i - 1] = "*"
    for row in array:
        print(' '.join(row))
```

Пропишем проверку на то, что наш скрипт выполняется напрямую (подробнее см. 3.1).

```python
if __name__ == "__main__":
    main()
```
