В данной задаче мы создаём пустое множество, далее вводим поступающее количество запросов N и реализуем эти запросы через консоль.
Для осуществления выполнения конкретного количества запросов N запустим цикл for от 0 до N. Внутри цикла принимаем через клавиатуру заданный запрос, и в зависимости от введённого запроса выполняем следующие алгоритмы:
При запросе ADD <число> - добавляем указанное число в множество,
При запросе COUNT - выводим в консоль количество элементов в множество,
При запросе PRESENT <число> - выводим в консоль ответ на вопрос, присутствует ли указанное число в множестве.
Такая программа помогает визуализировать свойства множеств - с её помощью мы можем наглядно увидеть, что при добавлении, например, одинаковых чисел, количество элементов в множестве останется равным количеству единичных экземпляров чисел, а через проверку PRESENT мы можем увидеть, что элемент действительно добавлен во множество.

Также примечательно, что в коде мы используем ранее не встречавшуюся в решениях функцию add() для множеств - она добавляет элемент, указанный в параметре, в начало множества.
Укажем также некоторые другие функции для множеств:
Очистка множества - clear()
Удаление элемента по индексу - remove()
Удаление последнего элемента и возврат - pop()
Проверка между двумя множествами на подмножество - issubset()
Проверка на надмножество - issuperset()

```python
def main():
    my_set = set()
    N = int(input().strip())
    for _ in range(N):
        query = input().strip().split()
        if query[0] == 'ADD':
            my_set.add(int(query[1]))
        elif query[0] == 'PRESENT':
            num = int(query[1])
            if num in my_set:
                print("YES")
            else:
                print("NO")
        elif query[0] == 'COUNT':
            print(len(my_set))
if __name__ == "__main__":
    main()
```
