Словари - следующая структура в Python, способная сохранять наборы данных. Визуально они идентичны множествам в плане использования фигурных скобок для инициализации, но имеют отличие в виде сохранения пар "ключ-значение", в то время как множества хранят внутри только сами значения объектов.
Словарь может быть инициализирован либо через функцию dict() (пример: dict(One = "1", Two = "2")), либо через присваивание переменной самого словаря (пример: {"One": '1', "Two": '2'}). Получение данных из словаря работает через указание ключа в квадратных скобках.
Примечательно, что в рамках нашей задачи также используем функцию get() для словарей - она может принять два параметра - первым ключ словаря, значение для которого мы хотим извлечь из функции, а вторым значение, которое вернёт функция при отсутствии указанного ключа. У нас данная функция встречается в восемнадцатой строке, где мы присваиваем по значению ключа элементу словаря сумму средств уже учтённых в словаре (либо же 0) и добавляемую вновь сумму (в случае 0 - впервые).
В общем наша программа состоит из инициализации словаря, последующего ввода через клавиатуру числа, равного количеству требуемых запросов, поле чего мы вводим сами запросы по стандартным конструкциям, и реализуем либо изменение в словаре суммы счёта клиента, либо вывод его актуальной суммы на балансе.

```python
def main():

    client_accounts = {}
    N = int(input().strip())
    for _ in range(N):
        query = input().strip().split()

        if query[0] == '1':
            name = query[1]
            amount = int(query[2])
            client_accounts[name] = client_accounts.get(name, 0) + amount

        elif query[0] == '2':
            name = query[1]
            if name in client_accounts:
                print(client_accounts[name])
            else:
                print('ERROR')

if __name__ == "__main__":
    main()
```

Выполненная задача:

У банка есть клиенты. Каждый клиент имеет ровно один счет.

Напишите программу (обязательно с использованием ассоциативного массива), которая будет выполнять последовательность
запросов таких двух видов:

1) начинается с числа 1, затем через пробел следует имя клиента (слово из латинских букв), далее через пробел идет сумма
денег, которая прилагается к счету текущего клиента (целое число, не превышает по модулю 10000).

2) начинается с числа 2, через пробел следует имя клиента. На каждый такой запрос программа должна ответить какая сумма
в данный момент есть на счету заданного клиента. Если такое имя клиента пока ни разу не упоминалось в запросах вида 1,
выводите вместо числа слово ERROR.

В начале работы программы у всех клиентов на счету 0. Затем суммы могут становиться как положительными, так и
отрицательными.

Обратите внимание, что в ситуации, когда клиент снял суммарно ровно столько же денег, сколько положил, сумма на счете
становится равной 0; но, раз его имя уже встречалось, нулевое значение не является основанием выводить ERROR.

Входные данные
Первая строка стандартного входного потока — количество запросов N (1<N<100000). Далее следуют N строк
в каждой из которых описан один из двух описанных выше видов запроса.

Выходные данные
На каждый запрос 2-го вида нужно вывести текущее значение на счету заданного клиента (или слово ERROR).
